## 3) check dataset is balanced or not

## make dataset balanced

```python
## synthetic minority oversampling technique 
##here it find k neighbours
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
x = df.drop('Class',axis=1)
y = df['Class']
```

```python
print('original dataset shape', Counter(y))
```

```python
y
```

```python
smote = SMOTE(random_state=42)
x_resampled,y_resampled = smote.fit_resample(x,y)
```

```python
print("resampled dataset shape", Counter(y_resampled))
```

```python
y_resampled
```

```python
x_resampled
```

```python
new_x = pd.DataFrame(x_resampled)
new_y = pd.DataFrame(y_resampled)
new_df = pd.concat([new_x,new_y],axis=1)
```

```python
new_df
```

```python
new_df['Class'].value_counts()
```

## 4) exploratory data analysis

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
new_df['numeric'] = label_encoder.fit_transform(new_df['Class'])
class_mapping = dict(zip(label_encoder.classes_,label_encoder.transform(label_encoder.classes_)))
print(class_mapping)
corr_df = new_df.drop('Class',axis=1)
corr =  corr_df.corr()
print(corr)
```

```python
threshold=0.3
high_corr_features = corr.index[abs(corr['numeric']) > threshold].tolist()
```

```python
high_corr_features
```

```python
import seaborn as sns
high_corr_features.append('Class')  # Add 'Class' back for hue parameter
sns.pairplot(new_df[high_corr_features], hue='Class',diag_kind='kde')
plt.show()
```

## 5) Data Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
print("shape of x_train:",x_train.shape)
print("shape of x_test:",x_test.shape)
print("shape of y_train:",y_train.shape)
print("shape of y_test:",y_test.shape)
```

## 6) Model Evaluation

a) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

from sklearn.model_selection import GridSearchCV
parameter = {'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
classifier_regressor = GridSearchCV(model,param_grid=parameter,scoring = 'accuracy',cv=5)
```

```python
classifier_regressor.fit(x_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
##prediction
y_pred = classifier_regressor.predict(x_test)
```

```python
## accuracy score
from sklearn.metrics import accuracy_score,classification_report
score = accuracy_score(y_pred, y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

b) KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=3
neigh = KNeighborsClassifier(n_neighbors=k).fit(x_train,y_train)
```

```python
neigh
```

```python
y_predict = neigh.predict(x_test)
```

```python
from sklearn import metrics
print("accuracy",metrics.accuracy_score(y_train,neigh.predict(x_train))*100)
```

c) SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)
print("model accuracy:{0:0.3f}".format(accuracy_score(y_test,y_pred)*100))
```

d) Decision Tree
